```python
import numpy as np

# Convert empty strings to NaN
df['rating'].replace('', np.nan, inplace=True)
df['discount_percentage'].replace('', np.nan, inplace=True)

# Fill NaN values with 0.0
df['rating'].fillna(0.0, inplace=True)
df['discount_percentage'].fillna(0.0, inplace=True)

# Convert to float
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Check if there are still issues
print(df[['rating', 'discount_percentage']].info())
```

```python
correlation = df['discount_percentage'].corr(df['rating'])
print(f"Correlation between discount_percentage and rating: {correlation:.3f}")
```

```python
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float).fillna(0).astype(int)
top_categories = df.groupby('category')['rating_count'].sum().sort_values(ascending=False)
print(top_categories.head(10))  # Display top 10 selling categories
```

```python
# Clean and convert 'discount_percentage' (remove % sign and convert to float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace(r'[^0-9.]', '', regex=True).astype(float)

# Clean and convert 'rating_count' (remove commas and convert to integer)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float).fillna(0).astype(int)

correlation = df['discount_percentage'].corr(df['rating_count'])
print(f"Correlation between discount and rating_count: {correlation:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort products by rating in descending order and pick the top 10
top_rated_products = df[['product_name', 'rating']].dropna().sort_values(by='rating', ascending=False).head(10)

# Set figure size
plt.figure(figsize=(12, 8))

# Create bar plot
sns.barplot(x='rating', y='product_name', data=top_rated_products, palette='viridis')

# Labels and title
plt.xlabel("Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Highest-Rated Products")
plt.xlim(0, 5)  # Assuming rating scale is 0-5

# Display plot
plt.show()

```

```python
import numpy as np

# Convert empty strings to NaN
df['rating'] = df['rating'].replace('', np.nan)
df['discount_percentage'] = df['discount_percentage'].replace('', np.nan)

# Fill missing values with 0.0 (or you can use mean/median if needed)
df['rating'] = df['rating'].astype(float).fillna(0.0)
df['discount_percentage'] = df['discount_percentage'].astype(float).fillna(0.0)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['discount_percentage'], y=df['rating'], alpha=0.6, color="blue")

plt.xlabel("Discount Percentage (%)")
plt.ylabel("Rating")
plt.title("Scatter Plot of Discount Percentage vs. Rating")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by category and sum the rating_count (assuming it represents sales)
top_categories = df.groupby('category')['rating_count'].sum().nlargest(5)  # Top 5 categories

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple', 'orange'])

plt.title("Top-Selling Categories")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```
